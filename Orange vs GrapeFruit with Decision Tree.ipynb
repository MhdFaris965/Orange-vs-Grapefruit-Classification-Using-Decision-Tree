```python
X=df.drop('name',axis=1)
y=df['name']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25)
```

```python
len(X)
```

```python
#sample training were will know the input and output
len(X_train)
```

```python
#outside sample only know the input
len(X_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtc=DecisionTreeClassifier()
dtc.fit(X_train,y_train)
```

```python
y_pred_train = dtc.predict(X_train)
y_pred_test = dtc.predict(X_test)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_train,y_pred_train))
```

```python
print(classification_report(y_test,y_pred_test))
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
hyper_para_dic = {'criterion':['gini','entropy'],
                  'max_depth':[2,4,6,8,10],
                  'min_samples_split':[2,3,5,7,9],
                  'min_samples_leaf':[2,4,5,8,9],
                  'max_features':['sqrt','log2']
                  }
```

```python
grid_model=GridSearchCV(estimator=dtc,param_grid=hyper_para_dic,cv=5)
```

```python
grid_model.fit(X_train,y_train)
```

```python
grid_model.best_params_
```

```python
best_dtc = grid_model.best_estimator_
```

```python
y_pred_train = best_dtc.predict(X_train)
y_pred_test = best_dtc.predict(X_test)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_train,y_pred_train))
```

```python
print(classification_report(y_test,y_pred_test))
```
